```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import time
import os
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
from selenium.common.exceptions import StaleElementReferenceException
```

```python
# Selenium setting
options = Options()
options.add_argument('--headless')  # run in background
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# WebDriver initialize
driver = webdriver.Chrome(options=options)
```

```python
# Create a folder to save images
save_folder_path = 'your_path'
if not os.path.exists(save_folder_path):
    os.makedirs(save_folder_path)
```

```python
def download_image(image_url, save_path):
    response = requests.get(image_url)
    if response.status_code == 200:
        with open(save_path, 'wb') as file:
            file.write(response.content)
```

```python
def scrape_and_move(start_url, save_folder_path):
    driver.get(start_url)
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Find and save the image URL
    images = soup.find_all('figure', {'class': 'comic-image-container'})  # The class names are examples. You will need to adjust it to fit your actual site.

    base_url = "https://omq.mo.cloudinary.net/_STRIPS/"

    for image in images:
        img_tag = image.find('img')
        if img_tag and 'src' in img_tag.attrs:
            img_url = img_tag['src']
            # If img_url is relative, prepend the base URL
            if not img_url.startswith(('http://', 'https://')):
                img_url = base_url + img_url
            print(img_url)  # check
            # Extract image file name (E.G: '20000113.gif')
            img_filename = img_url.split('/')[-1]
            # Set image storage path
            save_path = f'path/Q_comics/{img_filename}'  # Specify storage path
            # Download and save images
            download_image(img_url, save_path)


    # Set maximum number of attempts
    max_attempts = 3
    attempts = 0

    while attempts < max_attempts:
        try:
            # Find the link to the next page
            next_button = driver.find_element(By.XPATH, '//a[contains(@class, "neighbor-next") and @title="Next comic strip"]')
            next_page_url = next_button.get_attribute('href')
            print(next_page_url)
            start_url = next_page_url
            break  # The URL has been successfully fetched, so the loop ends.
        except StaleElementReferenceException:
            print("StaleElementReferenceException occurred, try again...")
            time.sleep(1)  # Wait a moment and try again
            attempts += 1

    # Click on the next page link
    if next_button:
        next_button.click()
        time.sleep(3)  # Wait for the next page to load
    return start_url
```

```python
# start URL
start_url = 'https://www.oldmasterq.com/comics/2496/'

# start scrapping
num_img = 100
for i in range(num_img):
  scrape_and_move(start_url, save_folder_path)
  start_url = scrape_and_move(start_url, save_folder_path)
```
