```python
from PIL import Image

# Load the images
image_path1 = 'your_path'
#image_path2 = '/mnt/data/D0690.jpg'

# Open the images using PIL
image1 = Image.open(image_path1)
#image2 = Image.open(image_path2)

# Function to crop the image into the specified number of cuts
def crop_image(image, cuts):
    width, height = image.size
    crop_height = height // cuts
    images_cropped = [image.crop((0, i * crop_height, width, (i + 1) * crop_height)) for i in range(cuts)]
    return images_cropped

# Crop the images
cropped_images_4cuts = crop_image(image1, 4)

# Save the cropped images
cropped_image_paths_4cuts = []

for i, img in enumerate(cropped_images_4cuts):
    path = f'your_path/image1_cropped_{i+1}.jpg'
    img.save(path)
    cropped_image_paths_4cuts.append(path)


```
